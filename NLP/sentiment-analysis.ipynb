```python
new_df = pd.DataFrame(columns=["region", "impressions_per_state"])
for row in range(negative_ads.shape[0]):
    temp_df = pd.DataFrame(negative_ads.get('delivery_by_region').iloc[row])
    temp_df['percentage'] = pd.to_numeric(temp_df['percentage'])
    temp_df['impressions_average'] = negative_ads.get('impressions_average').iloc[row]
    temp_df['impressions_per_state'] = temp_df['impressions_average'] * temp_df['percentage']
    new_df = pd.concat([new_df, temp_df], ignore_index = True, axis = 0)
    display(new_df)
    
```

```python
import altair as alt
import geopandas as gpd
from vega_datasets import data
import json
```

```python
states = alt.topo_feature(data.us_10m.url, feature='states')

alt.Chart(states).mark_geoshape(
    fill='lightgrey',
    stroke='white'
).project('albersUsa').properties(
    width=500,
    height=300
)
```

```python
pop = final_df
pop = pop.reset_index()
pop = pop.rename(columns={"region": "state", "impressions_per_state": "impressions"})
pop['id']=pop.index+1
pop.head()
```

```python
states = alt.topo_feature(data.us_10m.url, 'states')

variable_list = ['impressions']

viz = alt.Chart(states).mark_geoshape().encode(
    color='impressions:Q'
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(pop, 'id', list(pop.columns))
).properties(
    width=500,
    height=300
).project(
    type='albersUsa'
)

outline = alt.Chart(states).mark_geoshape(stroke='grey', fillOpacity=0).project(
    type='albersUsa'
).properties(
    width=500,
    height=300,
    title = 'Impressions per State for Negative Vaccine Ads'
)

alt.layer(viz, outline)
```
