```python
f = open("Ads10000_vaccine", "rb")
ads = pickle.load(f)
f.close() 
```

```python
df = pd.json_normalize(ads)
df
```

```python
df["impressions.lower_bound"] = pd.to_numeric(df["impressions.lower_bound"])
df["impressions.upper_bound"] = pd.to_numeric(df["impressions.upper_bound"])
df.dtypes
```

```python
df['impressions_average'] = (df["impressions.lower_bound"] + df["impressions.upper_bound"])/2
df
```

```python
"""
1. CREATE NEW DataFrame
2. iterate thru each row of negative_ads
3. create a dataframe of 'delivery_by_region' for that row
4. have that dataframe have columns of ["state", "percentage", "avg_impressions", "impressions_per_state"]
    a. multiply percentage * avg_impressions to get 'impressions_per_state'
    b. negative_ads['avg_impressions'] = curr_row.get('impressions_average')
5. add to the new dataframe we created in step 1:
    a. a row of the state and impressions_per_state
6. after iterating thru all rows of negative_ads, groupby regions and average the impressions_per_state column
7. develop a visualization/map of states and impressions (of negative ads) per state
"""
```

```python
new_df = pd.DataFrame(columns=["region", "impressions_per_state"])
for row in range(df.shape[0]):
    temp_df = pd.DataFrame(df.get('delivery_by_region').iloc[row])
    temp_df['percentage'] = pd.to_numeric(temp_df['percentage'])
    temp_df['impressions_average'] = df.get('impressions_average').iloc[row]
    temp_df['impressions_per_state'] = temp_df['impressions_average'] * temp_df['percentage']
    new_df = pd.concat([new_df, temp_df], ignore_index = True, axis = 0)
    
```

```python
pop = final_df
pop = pop.reset_index()
pop = pop.rename(columns={"region": "state", "impressions_per_state": "impressions"})
pop['id']=pop.index+1
pop.head()
```

```python
states = alt.topo_feature(data.us_10m.url, 'states')

variable_list = ['impressions']

viz = alt.Chart(states).mark_geoshape().encode(
    color='impressions:Q'
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(pop, 'id', list(pop.columns))
).properties(
    width=500,
    height=300
).project(
    type='albersUsa'
)

outline = alt.Chart(states).mark_geoshape(stroke='grey', fillOpacity=0).project(
    type='albersUsa'
).properties(
    width=500,
    height=300,
    title = 'Impressions per State for All Vaccine Ads'
)

alt.layer(viz, outline)
```
