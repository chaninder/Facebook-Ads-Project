```python
new_df = pd.DataFrame(columns=["age", "impressions_per_age_group"])
for row in range(df.shape[0]):
    temp_df = pd.DataFrame(df.get('demographic_distribution').iloc[row])
    temp_df['percentage'] = pd.to_numeric(temp_df['percentage'])
    temp_df['impressions_average'] = df.get('impressions_average').iloc[row]
    temp_df['impressions_per_age_group'] = temp_df['impressions_average'] * temp_df['percentage']
    temp_df=temp_df.filter(['age', 'impressions_per_age_group'])
    new_df = pd.concat([new_df, temp_df], ignore_index = True, axis = 0)
    #display(new_df)
    
```

```python
final=new_df.groupby(['age']).sum()
final
```

```python
col_total=final['impressions_per_age_group'].sum()
```

```python
final['proportion'] = final['impressions_per_age_group']/col_total
final=final.filter(['proportion'])
final
```

```python
age_groups = final.get('age').to_numpy()
data = final.get('proportion').to_numpy()

# Creating explode data
explode = (0.1, 0.0, 0.2, 0.1, 0.0, 0.0, 0.1)

# Creating color parameters
colors = ( "#003f5c", "#374c80", "#7a5195",
        "#bc5090", "#ef5675", "#ff764a", "#ffa600")

# Wedge properties
wp = { 'linewidth' : 0, 'edgecolor' : "white" }

# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
    autopct = lambda pct: func(pct, data),
    explode = explode,
    labels = age_groups,
    shadow = True,
    colors = colors,
    startangle = 90,
    wedgeprops = wp,
    textprops = dict(color ="white"))

# Adding legend
ax.legend(wedges, age_groups,
    title ="Age Groups",
    loc ="center left",
    bbox_to_anchor =(1, 0, 0.5, 1))

plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Average Impressions per Age Group (for all Vaccine ads)")

# show plot
plt.show()

```
